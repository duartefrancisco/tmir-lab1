```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def leerImagen(ruta):
    imagen = cv2.imread(ruta)

    return cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
```

```python
def escribirImagen(ruta, imagen):
    cv2.imwrite(ruta, imagen)
```

```python
def imgCanalesActivos(imagen, color):
    alto = imagen.shape[0]
    ancho = imagen.shape[1]

    imgProcesada = np.zeros((alto, ancho, 3))

    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = imagen[i,j]

            azul = pixel[2]
            verde = pixel[1]
            rojo = pixel[0]

            imgProcesada[i,j] = [azul if color == 1 or color == 20 or color == 30 else 0, 
                                verde if color == 2 or color == 10 or color == 20 else 0, 
                                rojo if color == 3 or color == 10 or color == 30 else 0]

    
    return imgProcesada
```

```python
img = leerImagen("./imagenes/leon.jpeg")
plt.imshow(img)
```

```python
escribirImagen("./imagenes/leon_salida.jpeg", imgCanalesActivos(img, 30))
```

```python
imgProcesada = leerImagen("./imagenes/leon_salida.jpeg")
plt.imshow(imgProcesada)
```
