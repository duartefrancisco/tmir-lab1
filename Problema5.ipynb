```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def leerImagen(ruta):
    imagen = cv2.imread(ruta)

    return cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
```

```python
def escribirImagen(ruta, imagen):
    cv2.imwrite(ruta, imagen)
```

```python
def imgEscalaGrises(imagen):
    alto = imagen.shape[0]
    ancho = imagen.shape[1]

    imgProcesada = np.zeros((alto, ancho, 1))

    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = imagen[i,j]

            azul = pixel[2]
            verde = pixel[1]
            rojo = pixel[0]

            imgProcesada[i,j] = int(0.299 * azul + 0.587 * verde + 0.114 * rojo)

    
    return imgProcesada
```

```python
img = leerImagen("./imagenes/leon.jpeg")
plt.imshow(img)
```

```python
escribirImagen("./imagenes/leon_salida_p5.jpeg", imgEscalaGrises(img))
```

```python
imgProcesada = leerImagen("./imagenes/leon_salida_p5.jpeg")
plt.imshow(imgProcesada)
```
