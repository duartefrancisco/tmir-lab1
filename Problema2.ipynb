```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def leerImagen(ruta):
    imagen = cv2.imread(ruta)

    return cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
```

```python
def escribirImagen(ruta, imagen):
    cv2.imwrite(ruta, imagen)
```

```python
def imgConvertirAColor(imgAzul, imgVerde, imgRojo):
    alto = imgAzul.shape[0]
    ancho = imgAzul.shape[1]

    imgProcesada = np.zeros((alto, ancho, 3))

    for i in range(0, alto):
        for j in range(0, ancho):

            azul = imgAzul[i,j][2]
            verde = imgVerde[i,j][1]
            rojo = imgRojo[i,j][0]

            imgProcesada[i,j] = [azul, verde, rojo]

    
    return imgProcesada
```

### Imagen 1

```python
img1Azul = leerImagen("./imagen1/imagen1_salida_gray_azul.jpg")
plt.imshow(img1Azul)
```

```python
img1Verde = leerImagen("./imagen1/imagen1_salida_gray_verde.jpg")
plt.imshow(img1Verde)
```

```python
img1Rojo = leerImagen("./imagen1/imagen1_salida_gray_rojo.jpg")
plt.imshow(img1Rojo)
```

```python
escribirImagen("./imagenes/imagen1_salida.jpeg", imgConvertirAColor(img1Azul, img1Verde, img1Rojo))
```

```python
img1Procesada = leerImagen("./imagenes/imagen1_salida.jpeg")
plt.imshow(img1Procesada)
```
