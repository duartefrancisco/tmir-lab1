```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def leerImagen(ruta):
    imagen = cv2.imread(ruta)

    return cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
```

```python
def escribirImagen(ruta, imagen):
    cv2.imwrite(ruta, imagen)
```

```python
def imgEscalaGrises(imagen, canal):
    alto = imagen.shape[0]
    ancho = imagen.shape[1]

    imgProcesada = np.zeros((alto, ancho, 1))

    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = imagen[i,j]

            azul = pixel[2] if canal == 1 else 0
            verde = pixel[1] if canal == 2 else 0
            rojo = pixel[0] if canal == 3 else 0

            imgProcesada[i,j] = int(0.299 * azul + 0.587 * verde + 0.114 * rojo)

    
    return imgProcesada
```

```python
img = leerImagen("./imagenes/leon2.jpg")
plt.imshow(img)
```

```python
escribirImagen("./imagenes/leon2_salida_azul.jpg", imgEscalaGrises(img,1))
```

```python
escribirImagen("./imagenes/leon2_salida_verde.jpg", imgEscalaGrises(img,2))
```

```python
escribirImagen("./imagenes/leon2_salida_rojo.jpg", imgEscalaGrises(img,3))
```
